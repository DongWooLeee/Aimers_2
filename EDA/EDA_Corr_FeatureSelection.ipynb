```python
import pandas as pd

df=train_df

# Compute the correlation matrix
corr_matrix = df.corr().abs()

# Set a threshold for identifying highly correlated columns
threshold = 0.8

# Find index of feature columns with correlation greater than threshold
# Exclude diagonal elements by using mask and np.ones_like function
mask = np.ones_like(corr_matrix, dtype=bool)
mask[np.triu_indices_from(mask)] = False
high_corr_cols = np.where(corr_matrix > threshold, mask, False)
high_corr_cols = np.column_stack(np.where(high_corr_cols))

# Identify the column(s) to keep from each pair of highly correlated columns
cols_to_drop = set()
for i, j in high_corr_cols:
    if i not in cols_to_drop and j not in cols_to_drop:
        corr_i = corr_matrix.iloc[i, i+1:]
        corr_j = corr_matrix.iloc[j, j+1:]
        if corr_i.mean() > corr_j.mean():
            cols_to_drop.add(j)
        else:
            cols_to_drop.add(i)

# Drop the highly correlated columns from the DataFrame
df.drop(df.columns[list(cols_to_drop)], axis=1, inplace=True)
train_df=df
```

# 상관계수 0.8 이상인 것 지우기

```python
import pandas as pd

# Load your dataset into a pandas DataFrame
df=train_df
# Compute the correlation matrix
corr_matrix = df.corr().abs()

# Set a threshold for identifying highly correlated columns
threshold = 0.8

# Find index of feature columns with correlation greater than threshold
# Exclude diagonal elements by using mask and np.ones_like function
mask = np.ones_like(corr_matrix, dtype=bool)
mask[np.triu_indices_from(mask)] = False
high_corr_cols = np.where(corr_matrix > threshold, mask, False)
high_corr_cols = np.column_stack(np.where(high_corr_cols))

# Identify the column(s) to keep from each pair of highly correlated columns
cols_to_drop = set()
for i, j in high_corr_cols:
    if i not in cols_to_drop and j not in cols_to_drop:
        corr_i = corr_matrix.iloc[i, i+1:]
        corr_j = corr_matrix.iloc[j, j+1:]
        if corr_i.mean() > corr_j.mean():
            cols_to_drop.add(j)
        else:
            cols_to_drop.add(i)

# Drop the highly correlated columns from the DataFrame
df.drop(df.columns[list(cols_to_drop)], axis=1, inplace=True)
```
